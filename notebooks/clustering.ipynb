# Clustering K-Means

```python
# import libraries
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('merged_train.csv', sep = ';', decimal = ',')
df.head()
```

```python
df.describe()
df.info()
```

```python
wss=[]

for i in range (1,11):
    kmeans = KMeans(n_clusters = i, max_iter = 300)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```
